## Performing EDA on the dataset 

### Understanding the dataset
- Shape
- Column names, datatypes
- Statistics: std, mean, count

```python
def dataset_info(dataset):

    dataset.info()

dataset_info(car_accidents)
```

```python
# Knowing the shape,columns,rows, column names and datatypes 
def get_dataset_shape(dataset):
    """
    Get the shape of the dataset

    Parameters:
    Pandas datafram

    Return 
    Tuple that shows the shape of the dataset 

    """

    shape = dataset.shape
    return shape
```

```python
car_accident_shape = get_dataset_shape(car_accidents)
car_accident_shape
```

```python
# Column names and types 
def get_column_info(dataset):

    """
    Get columns and the data types for each

    """

    column_info = pd.DataFrame({'Column Name' : dataset.columns, 'Data Type': dataset.dtypes})
    return column_info
```

```python
column_info_accidents = get_column_info(car_accidents)
print(column_info_accidents)
```

```python
# Statistics 
def get_statistics(dataset):

    statistics_values = dataset.describe().transpose()
    return statistics_values

statistics_car_accidents = get_statistics(car_accidents)
statistics_car_accidents
```

## Data Cleaning 

```python
# Check for duplicates 
def get_duplicates(dataset):

    duplicate_values = dataset.duplicated()
    return duplicate_values
```

```python
accident_duplicate_values = get_duplicates(car_accidents)
print(accident_duplicate_values)
```

```python
# Identify missing values 
def identify_missing_values(dataset):
    """
    Identify missing values in a dataset.
    
    Parameters:
    - dataset: Pandas DataFrame or Series
    
    Returns:
    - DataFrame with column names and the corresponding count of missing values
    """
    missing_values = dataset.isnull().sum().reset_index()
    missing_values.columns = ['Column Name', 'Missing Values']
    return missing_values
```

```python
missing_values_accidents = identify_missing_values(car_accidents)
print(missing_values_accidents)
```
